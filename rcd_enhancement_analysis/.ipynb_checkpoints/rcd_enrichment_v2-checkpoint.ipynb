```python
def add_xy_coords_metric(sdf, recenter=False):
    # Convert lat/long to x-y for easier distance calculations on kd-tree
    RE = 6378100 # radius of the earth in meters
    
    lon_rad = sdf["longitude"]*scipy.pi/180
    lat_rad = sdf["latitude"]*scipy.pi/180
    
    #small angle approximation: sin(theta) ~ theta if theta is very small
    x = RE*lon_rad * np.cos(np.mean(lat_rad))
    y = RE*lat_rad 
    
    if recenter:
        sdf["x"] = x-np.mean(x)
        sdf["y"] = y-np.mean(y)
    else:
        sdf["x"] = x
        sdf["y"] = y
    return sdf
    # d = R * sqrt( dx**2 + dy**2 )
    
```

```python
rbins_default = [5, 50, 100, 140, 500, 1000, 3000]


def get_positivity_at_radii(sdf_with_xy,
                            kdt,
                            rbins=rbins_default,
                            savefile=True,
                            savename="test.csv"):
    # sdf_with_xy: dataframe of people with rdt_pos, and x-y coordinates in meters
    # kdt: KDTree object
    pos_cut = sdf_with_xy["rdt_pos"] == 1
    num_pos = np.sum(pos_cut)

    pos_individuals_xy = np.array(sdf_with_xy[["x", "y"]][pos_cut])
    iid_list = list(sdf_with_xy["person_id"][pos_cut])
    rdt_pos = np.array(sdf_with_xy["rdt_pos"])

    # Data to be saved to CSV
    save_id_list = []
    save_r_list = []
    # cumulative
    save_n_c_list = []
    save_np_c_list = []
    save_p_c_list = []
    # differential
    save_n_d_list = []
    save_np_d_list = []
    save_p_d_list = []

    for ii in np.arange(num_pos):
        if ii % 1000 == 0:
            print(ii)

        X = pos_individuals_xy[ii]
        iid = iid_list[ii]

        for ri in range(len(rbins)):
            r = rbins[ri]

            pos, num_near_pts = positivity_within_radius(X, kdt, rdt_pos, r)
            save_id_list.append(iid)
            save_r_list.append(r)
            # cumulative
            save_n_c_list.append(num_near_pts)
            save_np_c_list.append(num_near_pts * pos)
            save_p_c_list.append(pos)

            #differential
            if ri == 0:
                save_n_d_list.append(num_near_pts)
                save_np_d_list.append(num_near_pts * pos)
                save_p_d_list.append(pos)
            else:
                n_d = num_near_pts-save_n_c_list[-2]
                np_d = num_near_pts * pos-save_np_c_list[-2]
                save_n_d_list.append(n_d)
                save_np_d_list.append(np_d)
                save_p_d_list.append(np_d/n_d)

    return_df = pd.DataFrame({
        "person_id": save_id_list,
        "r": save_r_list,
        "n_cum": save_n_c_list,
        "np_cum": save_np_c_list,
        "prev_cum": save_p_c_list, 
        "n_diff": save_n_d_list,
        "np_diff": save_np_d_list,
        "prev_diff": save_p_d_list
    })

    if savefile:
        return_df.to_csv(savename, index=False)

    return return_df
```

```python
def return_trend(x, y, bins, weights=None, weighted=False):
    bins_centered = (bins[1:]+bins[:-1])/2.
    
    trend = []
    for i in range(len(bins)-1):
        cut = np.logical_and(x > bins[i], x <= bins[i+1])
        if weighted:
            center = np.average(y[cut], weights=weights[cut])
        else:
            center = np.average(y[cut])
        trend.append(center)
    
    trend = np.array(trend)
    
    return bins_centered, trend
```

```python
def calculate_avg_prev_around_index_cases(df, ckdt, rmin=0,rmax=5):
    pos_cut = df["rdt_pos"] == 1
    num_pos = np.sum(pos_cut)
    pos_individuals_xy = np.array(df[["x", "y"]][pos_cut])
    rdt_pos = np.array(df["rdt_pos"])
    
    # Loop over every RDT-positive, and calculate prevalence around this person within rmin and rmax.
    
    prev_array = np.zeros(num_pos)
    tot_n_tracker = 0
    tot_np_tracker = 0
    for ii in np.arange(num_pos):
        X = pos_individuals_xy[ii]
        
        pos, num_near_pts = positivity_within_radius(X, ckdt, rdt_pos, rmax, rmin=rmin)
        prev_array[ii] = pos
        
        print(pos)
        print(num_near_pts)
        tot_n_tracker += num_near_pts
        tot_np_tracker += num_near_pts*pos
        
#     print(tot_n_tracker)
#     print(tot_np_tracker)
        
    prev_array_without_nans = prev_array[np.logical_not(np.isnan(prev_array))]
    return np.mean(prev_array_without_nans)
    
def bootstrap_permutations(df, rmin=0, rmax=5, n_permutations=999):
    pass
```

```python
def positivity_within_radius(X, kdt, rdt_pos, rmax, rmin=0):
    near_pts = kdt.query_ball_point(X,rmax)
    
    print(len(near_pts))
    if rmin > 0:
        too_near_pts = kdt.query_ball_point(X,rmin)
        # Remove points that are too-near from near_pts
        exclude_too_near = np.isin(near_pts,too_near_pts, invert=True)
        print("exclude_too_near ",exclude_too_near)
        print(len(too_near_pts))
        if np.sum(exclude_too_near) == 0:
            pass
        else:
            near_pts = near_pts[exclude_too_near] 
        
    num_near_pts = np.float(len(near_pts))

    if rmin == 0:
        pos = (np.sum(rdt_pos[near_pts])-1)/(num_near_pts-1) # minus one to not recount index person themselves
        num_near_pts -= 1
    elif num_near_pts == 0:
        pos = np.nan
        num_near_pts = 0
    else:
        pos = (np.sum(rdt_pos[near_pts]))/(num_near_pts)
    
    return pos, num_near_pts
```
