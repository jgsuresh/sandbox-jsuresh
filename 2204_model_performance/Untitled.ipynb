```python
# Can we separate decay and growth timesteps?
```

```python
tau = 0.5
boost_factor=10
```

```python
def exponential_decay(y, dt):
    return y*np.exp(-dt/tau)
```

```python
def boost(y, dt):
    return y + dt/tau * boost_factor
```

```python
def decay_then_boost(y,dt):
    y_decayed = exponential_decay(y,dt)
    return boost(y_decayed, dt)

def decay_and_boost_together(y,dt):
    return (y + dt/tau * boost_factor)*np.exp(-dt/tau)
```

```python
dt = 1/8
n_timesteps = 100
t = np.cumsum(np.ones(n_timesteps)*dt)

y_start = 0.5
for ti in t:
    if ti == dt:
        y_new = y_start
    y_new = decay_then_boost(y_new,dt)
    # y_new = decay_and_boost_together(y_new,dt)
    print(y_new)
```
