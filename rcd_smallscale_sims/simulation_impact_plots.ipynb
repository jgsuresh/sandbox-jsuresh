```python
def merge_and_rename_shared_column(df1, df2, column_to_merge_on, shared_column, rename_df1_column, rename_df2_column, how="inner"):
    columns_to_keep = [column_to_merge_on, shared_column]
    
    hold1 = df1[columns_to_keep].rename(columns={shared_column: rename_df1_column})
    hold2 = df2[columns_to_keep].rename(columns={shared_column: rename_df2_column})
    
    df_merged = hold1.merge(hold2, on=column_to_merge_on, how=how).reset_index(drop=True)
    
    return df_merged
    
    
```
