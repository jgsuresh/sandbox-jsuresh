## How do clinical cases compare to eachother

```python
pop_by_age_bin = data_sum['DataByTimeAndAgeBins']['Average Population by Age Bin']
clinical_incidence_by_age_bin = data_sum['DataByTimeAndAgeBins']['Annual Clinical Incidence by Age Bin']

clinical_incidence_summary = np.array([
    np.sum(np.array(clinical_incidence_by_age_bin[i])*np.array(pop_by_age_bin[i])) for i in [0,1,2,3]
])
```

```python
clinical_incidence_daily = np.array(data_inset['Channels']['New Clinical Cases']['Data'])
clinical_incidence_inset = np.array([np.sum(clinical_incidence_daily[i*365:(i+1)*365]) for i in [0,1,2,3]])
```

```python
print(clinical_incidence_inset/clinical_incidence_summary)
```

## How does RDT-prevalence compare to eachother

```python
pop_by_age_bin = data_sum['DataByTimeAndAgeBins']['Average Population by Age Bin']
rdtprev_by_age_bin = data_sum['DataByTimeAndAgeBins']['PfPR by Age Bin']

rdtprev_summary = np.array([
    np.sum(np.array(rdtprev_by_age_bin[i])*np.array(pop_by_age_bin[i]))/np.sum(pop_by_age_bin[i]) for i in [0,1,2,3]
])
```

```python
rdtprev_daily = np.array(data_inset['Channels']['Blood Smear Parasite Prevalence']['Data'])
rdtprev_inset = np.array([np.mean(rdtprev_daily[i*365:(i+1)*365]) for i in [0,1,2,3]])
```

```python
print(rdtprev_inset/rdtprev_summary)
```
