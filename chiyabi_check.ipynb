```python
from sim_map import *


def plot_best_by_round(exp_name, rnd, channel_list, n=5, ref_date=None, filenames=None, write_dir=None, ref_data_dict={}):
    df = pd.read_csv("prev_dist_by_round.csv")

    df = df[df["round"]==rnd]
    df.sort_values(by=["sim_dist"], ascending=True, inplace=True)
    
    am = AnalyzeManager()
    for sim_id in df["id"].iloc[0:n]:
        am.add_simulation(retrieve_simulation(sim_id))
    for channel in channel_list:
        if channel != "Monthly Incidence":
            if bool(ref_data_dict) and channel in ref_data_dict:
                ref = ref_data_dict[channel]
                am.add_analyzer(basic_inset_channel_plotter(channel, filenames=filenames, ref_date=ref_date, working_dir=write_dir, reference_data=ref, savefile_prefix="plot_best_{}".format(rnd)))
            else:
                am.add_analyzer(
                    basic_inset_channel_plotter(channel, filenames=filenames, ref_date=ref_date, working_dir=write_dir, savefile_prefix="plot_best_rd{}".format(rnd)))

        elif channel == "Monthly Incidence":
            if bool(ref_data_dict) and channel in ref_data_dict:
                ref = ref_data_dict[channel]
                am.add_analyzer(counter_plotter("Received_Treatment", ref_date=ref_date, reference_data=ref, savefile_prefix="plot_best_{}_".format(rnd)))
    am.analyze()

```

```python
if __name__=="__main__":
    plot_best_by_round(exp_id, 0, ["True Prevalence"])
```
