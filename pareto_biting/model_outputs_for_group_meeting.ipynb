```python
def get_mean_in_window(df, t_start, t_end, return_raw_in_window=False):
    foo = df[np.logical_and(df["timestep"] >= t_start, df["timestep"] < t_end)]
    bar = foo.groupby("timestep").agg("mean").reset_index()
    
    if return_raw_in_window:
        return bar, foo
    else:
        return bar
```

How many simulations eliminated?

```python
def sim_eliminated(sim_df):
    if np.sum(sim_df["RDT_prev"] == 0) > 365:
        return True
    else:
        return False
    
def percent_sims_eliminated(exp_df):
    eliminated=0
    for r, sim_df in exp_df.groupby("Run_Number"):
        eliminated += sim_eliminated(sim_df)
    return eliminated/10.
```

```python
year_start = 45
year_end = 50
u,uraw = get_mean_in_window(df_uniform_low,year_start*365,year_end*365, return_raw_in_window=True)
l,lraw = get_mean_in_window(df_lognormal_low,year_start*365,year_end*365, return_raw_in_window=True)
plt.scatter(uraw["timestep"], uraw["RDT_prev"], c="C0", s=0.1)
plt.scatter(lraw["timestep"], lraw["RDT_prev"], c="C1", s=0.1)
plt.plot(u["timestep"], u["RDT_prev"], label="Uniform")
plt.plot(l["timestep"], l["RDT_prev"], label="Lognormal")
plt.legend()

print(percent_sims_eliminated(uraw))
print(percent_sims_eliminated(lraw))
```
