```python
def get_mean_in_window(df, t_start, t_end, return_raw_in_window=False):
    foo = df[np.logical_and(df["timestep"] >= t_start, df["timestep"] < t_end)]
    bar = foo.groupby(["timestep","funest"]).agg("mean").reset_index()
    
    if return_raw_in_window:
        return bar, foo
    else:
        return bar
```

```python
def sim_eliminated(sim_df):
    end_df = sim_df[sim_df["timestep"]>=365*64]
    if np.sum(end_df["RDT_prev"]) == 0:
        return True
    else:
        return False
    
def percent_sims_eliminated(exp_df):
    eliminated=0
    nsim = 0
    for r, sim_df in exp_df.groupby("Run_Number"):
        nsim += 1
        eliminated += sim_eliminated(sim_df)
#     print(nsim)
#     print(eliminated)
    return eliminated/nsim
```

```python
percent_sims_eliminated(df_uniform)
percent_sims_eliminated(df_lognormal)
percent_sims_eliminated(df_exponential)
```

```python
eliminated = np.zeros(20)
f_arr = np.zeros(20)
i = 0
for f, sdf in df_uniform.groupby("funest"):
    eliminated[i] = percent_sims_eliminated(sdf)
    f_arr[i] = f
    i += 1

df_save1 = pd.DataFrame({"biting_type": "uniform", "funest": f_arr, "eliminated": eliminated})
```

```python
eliminated = np.zeros(20)
f_arr = np.zeros(20)
i = 0
for f, sdf in df_exponential.groupby("funest"):
    eliminated[i] = percent_sims_eliminated(sdf)
    f_arr[i] = f
    i += 1

df_save2 = pd.DataFrame({"biting_type": "exponential", "funest": f_arr, "eliminated": eliminated})
```

```python
eliminated = np.zeros(20)
f_arr = np.zeros(20)
i = 0
for f, sdf in df_lognormal.groupby("funest"):
    eliminated[i] = percent_sims_eliminated(sdf)
    f_arr[i] = f
    i += 1

df_save3 = pd.DataFrame({"biting_type": "lognormal", "funest": f_arr, "eliminated": eliminated})
```

```python
df_elim = pd.concat([df_save1,df_save2,df_save3])

df_elim = df_elim.merge(df_end, on="funest")
```

```python
plt.figure(dpi=200)
for b, sdf in df_elim.groupby("biting_type"):
    if b == "uniform":
        c = "C0"
    elif b == "exponential":
        c = "C1"
    elif b == "lognormal":
        c = "C2"
    plt.plot(sdf["daily_bite_rate"], sdf["eliminated"], marker='o',label=b,c=c)
plt.xlim([0.2,1])

plt.xlabel("Average daily bite rate")
plt.ylabel("Percent of sims that eliminate")
plt.legend()
```

# More detailed elimination sweep

```python
ln = pd.read_csv(input_folder + "endpoints_6e073cd6-6865-eb11-a2dd-c4346bcb7271.csv")
ex = pd.read_csv(input_folder + "endpoints_44bad60e-6965-eb11-a2dd-c4346bcb7271.csv")
un = pd.read_csv(input_folder + "endpoints_8ed2aa38-6965-eb11-a2dd-c4346bcb7271.csv")
```

```python
ln_new = pd.read_csv(input_folder + "endpoints_4bda73fd-f165-eb11-a2dd-c4346bcb7271.csv")
```

```python
ln_e = []
ex_e = []
un_e = []

for f, sdf in ln.groupby("funest"):
    ln_e.append(percent_sims_eliminated(sdf))
    
for f, sdf in ex.groupby("funest"):
    ex_e.append(percent_sims_eliminated(sdf))
    
for f, sdf in un.groupby("funest"):
    un_e.append(percent_sims_eliminated(sdf))

ln_e = np.array(ln_e)
ex_e = np.array(ex_e)
un_e = np.array(un_e)

bites_df = un.groupby("funest").agg({"daily_bite_rate": "mean"}).reset_index()
```

```python
ln_new_e = []

for f, sdf in ln_new.groupby("funest"):
    ln_new_e.append(percent_sims_eliminated(sdf))
    
ln_new_e = np.array(ln_new_e)

bites_df_2 = ln_new.groupby("funest").agg({"daily_bite_rate": "mean"}).reset_index()
```

```python
plt.figure(dpi=200)

x = np.array(bites_df["daily_bite_rate"])
plt.plot(x,un_e, marker='o', label="Uniform")
plt.plot(x,ex_e, marker='o', label="Exponential")
# plt.plot(x,ln_e, marker='o', label="Log-Normal")
plt.plot(np.array(bites_df_2["daily_bite_rate"]), ln_new_e, marker='o',label="Log-Normal")
plt.legend()

plt.xlabel("Average daily bites per person")
plt.ylabel("Fraction of sims that eliminate")
plt.xscale("log")
```
